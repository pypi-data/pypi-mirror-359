```python
from anarcii import Anarcii

model = Anarcii(
    seq_type="unknown",
    batch_size=128,
    cpu=False,
    ncpu=12,
    mode="accuracy",
    verbose=True, # By setting to True - when we renumber a PDB we can see the renumbered chains.
)
```

```python
pdb = "../tests/data/raw_data/1kb5.pdb"

# As we are renumbering a PDB file we can set the file path without .pdb (pdb_out_stem) 
# If this is not set the new file with _anarcii appended will appear in the working dir
results = model.number(pdb, pdb_out_stem="tmp/renumbered_1kb5")
```
